## Problem Statement 

***Analyze Diwali sales data to improve customer experience and sales***

```python
from pandas import*
from numpy import*
from matplotlib import pyplot as plt
import seaborn 
```

```python
#Import data
data=read_csv(r"Diwali Sales Data.csv",encoding="unicode escape")
```

```python
#shape return total num of rows and column
data.shape
```

```python
#head() return top first rows
data.head()
```

```python
#is return information about data present into table
data.info()
```

```python
# drop multiple columns at a time
data.drop(columns=["Status","unnamed1"],axis=1,inplace=True)
```

```python
#isnull used check null values present or not in that particular colum
data.isnull().sum()
```

```python
#remove null values from specific column
data.dropna(subset=["Amount"],inplace=True)
```

```python
data.isnull().sum()
```

```python
#describe used for statistical analysis of data
data[["Age","Orders","Amount"]].describe()
```

```python
# astype()- modify data type of particular column
data["Amount"]=data["Amount"].astype('int')
```

```python
#info is used to check datatype of that particular column
data.info()
```

```python
#columns is used to check column names
data.columns
```

```python
#rename is used rename column name

data.rename(columns={"Cust_name":"Customer_name"},inplace=True)
data.head()
```

```python
# total count male and female
```

```python
gender_count=data["Gender"].value_counts()

bars=plt.bar(gender_count.index,gender_count.values,color=["royalblue","y"],edgecolor="black")
plt.title("Gender vs count ",color="black")
plt.xlabel("Gender")
plt.ylabel("Count")
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        int(height),
        ha='center',
        va='bottom'
    )

plt.show()
```

***From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men***

```python
# male and female age groupwise total count of orders
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# create table
table = pd.crosstab(d['Age Group'], d['Gender'])

age_groups = table.index
male = table['M']
female = table['F']

x = np.arange(len(age_groups))
width = 0.35

plt.bar(x - width/2, male, width, label='Male')
plt.bar(x + width/2, female, width, label='Female')

plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Age Group-wise Gender Count (Grouped Bar Chart)")
plt.xticks(x, age_groups)

plt.legend()
plt.show()
```

***From above graphs we can see that most of the buyers are of age group between 26-35 yrs female***

```python
order_count=d.groupby(['State'],as_index=False)['Orders'].count().sort_values(by='Orders',ascending=False).head(10)
```

```python
plt.figure(figsize=(15,5))
plt.bar(order_count['State'],order_count['Orders'],data=order_count,color=["royalblue","gold","gray","cyan","lightgreen","lightcoral","lightblue",],edgecolor="black")
plt.xticks(rotation=45)


plt.xlabel("State")
plt.ylabel("No of order")
plt.title("State Vs No Order")

plt.show()
```
